```python
import pandas as pd
import requests
import json
```

```python
import os 
from dotenv import load_dotenv
load_dotenv()
API_KEY = os.getenv("My_API_Key")

def get_data():
    url = f"https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=EUR&apikey={API_KEY}"
    response = requests.get(url)
    values = response.json()
    print(values)  # This will print the whole response from the API
    return values

# Get the data from the API
data = get_data()
```

```python
# 'Time Series (Daily)' is the key where the date-based data is located in the API response
time_series = data.get("Time Series (Digital Currency Daily)")
time_series
```

```python
# Get the current date
current_date = pd.to_datetime('today')

# Filter the data to include only the last 6 months
six_months_ago = current_date - pd.DateOffset(months=6)

#Bitcoin_last_6_months = time_series[time_series['date'] >= six_months_ago]
```

```python
Bitcoin_Date = []
Bitcoin_Open = []
Bitcoin_High = []
Bitcoin_Low = []
Bitcoin_Close = []
Bitcoin_Volumn = []



for date1, daily_values in time_series.items():
    if (isinstance(daily_values, dict)) and ('1. open' in daily_values) and (pd.to_datetime(date1) > six_months_ago):
        Bitcoin_Date.append(pd.to_datetime(date1))
        Bitcoin_Open.append(float(daily_values["1. open"]))
        Bitcoin_High.append(float(daily_values['2. high']))
        Bitcoin_Low.append(float(daily_values['3. low']))
        Bitcoin_Close.append(float(daily_values['4. close']))
        Bitcoin_Volumn.append(float(daily_values['5. volume']))      
    # else:
    #     print(f"Unexpected structure for {date1}: {daily_values}")

print (Bitcoin_Date, Bitcoin_Open, Bitcoin_High, Bitcoin_Low, Bitcoin_Close, Bitcoin_Volumn)

# Reverse the dates and opening values if needed to get the chronological order
Bitcoin_Date.reverse()
Bitcoin_Open.reverse()
Bitcoin_High.reverse()
Bitcoin_Low.reverse()
Bitcoin_Close.reverse()
Bitcoin_Volumn.reverse()
```

```python
import psycopg

dbconn = os.getenv("DBCONN")
conn = psycopg.connect(dbconn)

# Create Cursor
Cur = conn.cursor()
```

```python
for i in range(0,len(Bitcoin_Date)):
    Cur.execute(
        '''
        INSERT INTO "Bitcoin_Data"("Date", "Open", "High", "Low", "Close", "Volumn")
        VALUES (%s, %s, %s, %s, %s, %s);    
        ''',
        (Bitcoin_Date[i], Bitcoin_Open[i], Bitcoin_High[i], Bitcoin_Low[i], Bitcoin_Close[i], Bitcoin_Volumn[i])    
    )
```

```python
# commit to all executed queries

conn.commit ()

# close the cursor and server Connection
Cur.close()
conn.close()
```
