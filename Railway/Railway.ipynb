```python
import pandas as pd
import matplotlib.pyplot as plt
from My_Key import Personal_Key
import requests
import json
```

```python
def get_data():
    url = f"https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=EUR&apikey=" + Personal_Key
    response = requests.get(url)
    values = response.json()
    print(values)  # This will print the whole response from the API
    return values

# Get the data from the API
data = get_data()
```

```python
# 'Time Series (Daily)' is the key where the date-based data is located in the API response
time_series = data.get("Time Series (Digital Currency Daily)")
time_series
```

```python
# Get the current date
current_date = pd.to_datetime('today')

# Filter the data to include only the last 6 months
six_months_ago = current_date - pd.DateOffset(months=6)

#Bitcoin_last_6_months = time_series[time_series['date'] >= six_months_ago]
```

```python
Bitcoin_Date = []
Bitcoin_Open = []
Bitcoin_High = []
Bitcoin_Low = []
Bitcoin_Close = []
Bitcoin_Volumn = []



for date1, daily_values in time_series.items():
    if (isinstance(daily_values, dict)) and ('1. open' in daily_values) and (pd.to_datetime(date1) > six_months_ago):
        Bitcoin_Date.append(pd.to_datetime(date1))
        Bitcoin_Open.append(float(daily_values["1. open"]))
        Bitcoin_High.append(float(daily_values['2. high']))
        Bitcoin_Low.append(float(daily_values['3. low']))
        Bitcoin_Close.append(float(daily_values['4. close']))
        Bitcoin_Volumn.append(float(daily_values['5. volume']))      
    # else:
    #     print(f"Unexpected structure for {date1}: {daily_values}")

print (Bitcoin_Date, Bitcoin_Open, Bitcoin_High, Bitcoin_Low, Bitcoin_Close, Bitcoin_Volumn)

# Reverse the dates and opening values if needed to get the chronological order
Bitcoin_Date.reverse()
Bitcoin_Open.reverse()
Bitcoin_High.reverse()
Bitcoin_Low.reverse()
Bitcoin_Close.reverse()
Bitcoin_Volumn.reverse()
```

```python
for i in range(0,len(Bitcoin_Date)):
    Cur.execute(
        '''
        INSERT INTO "Bitcoin_Data"("Date", "Open", "High", "Low", "Close", "Volumn")
        VALUES (%s, %s, %s, %s, %s, %s);    
        ''',
        (Bitcoin_Date[i], Bitcoin_Open[i], Bitcoin_High[i], Bitcoin_Low[i], Bitcoin_Close[i], Bitcoin_Volumn[i])    
    )
```

```python
## News scrapping

import pandas as pd
import seaborn as sns
import requests
import bs4
```

```python
response = requests.get("https://www.ft.com/search?sort=relevance&q=bitcoin")
soup = bs4.BeautifulSoup(response.text, 'html.parser') # without lxml package
soup
```

```python
Bitcoin_news = soup.select_one("div.search-results.o-teaser-collection")
Articles = Bitcoin_news.select("div.o-teaser__heading a")
Articles
Heading = []

for Head in Articles:
    #print (Head.get_text())
    Heading.append(Head.get_text())

print (Heading)
len(Heading)
```

```python
## Check final version for correted code 

N_Link = Bitcoin_news.select("div.o-teaser__heading a")
Link = []

for L in N_Link:
    Link.append(L["href"])

print(N_Link)
len(Link)
```

```python
Details1 = Bitcoin_news.select("p.o-teaser__standfirst a")
Details1

Detail = []

for det in Details1:
    Detail.append(det.get_text())

print (Detail)
len (Detail)
```

```python
Article_Catagory1 = Bitcoin_news.select("div.o-teaser__meta a")
print (Article_Catagory1)

print (len(Article_Catagory1))

Article_Catagory = []

for Dt1 in Article_Catagory1:
    #print (Dt1.get_text())
    #text1 = Dt1.get_text().strip()  # Strip any leading or trailing spaces
    text1 = Dt1['aria-label']
    Article_Catagory.append(text1)

print (Article_Catagory)
len(Article_Catagory)
```

```python
Date = Bitcoin_news.select("div.o-teaser__timestamp")
#print (Date)

Publish_Date = []

for Dt in Date:
    #print (Dt.get_text())
    text = Dt.get_text().strip()  # Strip any leading or trailing spaces
    Publish_Date.append(text)


print (Publish_Date)
len(Publish_Date)
```

```python
len(Article_Catagory)
len(Publish_Date)

# Expanding second_array to have 25 entries, by adding modified entries when 'Live blog' is present
for i in range(len(Article_Catagory)):
    if Article_Catagory[i].startswith("Live blog"):
        # If 'Live blog' is found, add the rest of the string (after removing 'Live blog') at the same index
        Publish_Date.insert(i, Article_Catagory[i].replace("Live blog from", "").strip())

# If second_array has less than 25 entries, fill it with some default value like "no_data"
while len(Publish_Date) < 25:
    Publish_Date.append("no_data")

print(Publish_Date)
len (Publish_Date)
```

```python
for i in range(0,24):
    Cur.execute(
        '''
        INSERT INTO "Bitcoin_News"("News_Heading", "News_Details", "Catagory_of_Article", "Date_of_Publication")
        VALUES (%s, %s, %s, %s);    
        ''',
        (Heading[i], Detail[i], Article_Catagory[i], Publish_Date[i])    
    )
```
